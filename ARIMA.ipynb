# Data preprocessing

```python
import numpy as np
import json

import pandas as pd
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
```

```python
with open("./data/carpark_imputed_nonull.json", "r") as f:
    dataset = json.load(f)
```

## Studying autocorrelation

```python
train_df = pd.DataFrame({"util":dataset["HE12"]["util_pct"]})
```

```python
autocorrelation_plot(train_df)
```

```python
plot_acf(train_df, lags=80, zero=False)
```

```python
plot_pacf(train_df, lags=80, zero=False)
```
